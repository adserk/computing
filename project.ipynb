##   <font color='Blue'>Introduction</font> 
<a id = 'introduction'></a>

The age of the universe is theorised to be the time elapsed since the Big Bang, where everything is considered to be at a singular point. It is an important area for cosmologists to study because by calculating for the age of the universe we can then learn more about the fate of the universe, whether it will end in a 'Big Crunch'(universe collapse) or 'Big Freeze' where the universe expands forever. Hence calculating the age is crucial to understand the cosmological theory of our universe.

There are different methods of measuring the age of our universe. One of the ways is to measure the Hubble constant, which is the rate of expansion of the universe.  By constructing a Hubble Diagram, which gives us the Hubble constant, we need the stars’ velocity, redshift and its relative distance. Measuring intergalactic distances will be difficult without objects with a known absolute magnitude M. These are called 'Standard Candles' (e.g. Type Ia supernovae). Type Ia supernovae are extremely valuable to determine large distances(1Mpc-1000Mpc). 1 Mpc = $3.086 \times 10^{19}$km.

The radiation is theorised to be the residual energy left from the Big Bang. Hence it is an important source of information for cosmologists to study the early universe.

## <font color='Blue'>Methods</font> 
<a id = 'methods'></a>

To find the age of the universe, we will use the integral form of Friedmann's equation to integrate for the area of the integrand. Also, we will be studying different models of the universe with varying density parameters $\Omega$. This will allow us to see the model which best describe our experimental results.
The Friedmann equations are a set of equations in cosmology that determines the expansion of the universe in a homogeneous and isotropic model of the universe. They were first derived by Alexander Friedmann in 1922 from Einstein's Field Equations of gravity. 
<img src="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSo8-BgBTJdq8Mb1HlxhpYQOwoMKeREHxndpy7b4J2o1C1RIp6F" width="150" height="150" />

__[Alexander Friedmann](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSo8-BgBTJdq8Mb1HlxhpYQOwoMKeREHxndpy7b4J2o1C1RIp6F)__




Equation 1: The Friedmann Equation 
$$ H^2 = \bigg(\frac{\dot{a}}{a}\bigg)^2 = \frac{8\pi G}{3} \bigg(\rho_M + \rho_R + \rho_\Lambda \bigg) -\frac{kc^2}{a^2} $$ 

$H$ = Hubble's constant, 
$G$ = Gravitational constant, 
$\rho_M$ = Energy density of matter,
$\rho_R$ = Energy density of Radiation,
$\rho_\Lambda$ = Energy density of Dark Energy,
$k$ = Curvature of the universe (+1,-1,0),
$a$ = scale factor, 
$\dot{a}$ = differential of $a$ with respect to time. 


k = 1 Closed universe
k = -1 Open universe
k = 0 Flat universe

Equation 1.1: Integral form of Friedmann Equation
$$ t_0 = \int_0^a \frac{da}{ a H_0 \bigg(\frac{\Omega_{k0}}{a^2} + \frac{\Omega_{M0}}{a^3} + \frac{\Omega_{R0}}{a^4} + \Omega_{\Lambda 0} \bigg)^{\frac{1}{2}}} $$

$t_0$ = age of the universe, 
$\Omega_{k0}$ = Density parameter of universe's curvature,
$\Omega_{M0}$ = Density parameter of matter, 
$\Omega_{R0}$ = Density parameter of Radiation,
$\Omega_{\Lambda 0}$ = Density parameter of Dark Energy. 
We will be using the integral form of Friedmann's equation to integrate and evaluate the age of the universe for different cosmological models.

- Note that the critical density parameters must add up to 1 for a flat universe $\Omega_0$ = 1 (i.e $\Omega_k + \Omega_M + \Omega_R + \Omega_\Lambda $ = 1)

However, when we try to evaluate for the Closed or Open universe we will require different equations to model the universe as we can only integrate up to the maximum point $a(t)$ = 1 using equation 1.1. This means we can never reach the full closed universe model.    
Ryden has written the equation which relates the curvature, density and the expansion rate in this form: 

Equation 1.2
$$\Omega_0 = 1 - \frac{kc^2}{R_0 ^2 H_0 ^2}$$
(Ryden 4.3). 
This suggests that if: 

1) k = - 1 then $\Omega_0$ < 1

2) k =   1 then $\Omega_0$ > 1

3) k =   0 then $\Omega_0$ = 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate #integration for the Simpson's method
# %matplotlib notebook
%config InlineBackend.figure_format = 'retina' #double resolution plot output for retina screens
```

```python
'''The constants used in units conversion when calculating time in Gyr '''
Mpc=3.086e22 #metres
km=1e3  #metres
Gyr=3.1536e16   #seconds
```

```python
'''Returns the integrand of Friedmann equation '''
def Friedmann(a,H,om_k,om_M,om_R,om_lambda):
    equation = 1/ ((a*H)*( (om_k/a**2) +(om_M/a**3) + (om_R/a**4) + (om_lambda))**0.5)
    return equation
```

### $\S$ 1.1 $\Lambda CDM $ Model  $(\Omega_M = 0.3 , \Omega_\Lambda = 0.7 )$
<a id = 'section1.1'></a>
The Lambda-CDM is a model of a universe which contains a cosmological constant $\Lambda$ associated with dark energy (vacuum energy) and Cold Dark Matter (CDM). This is also referred to as the 'standard model' of the Big Bang cosmology. It is noteworthy to say that this model assumes General Relativity is the correct theory of gravity on cosmological scales. 
According to this model, our universe contains matter (baryonic and dark), radiation and dark energy, whose combined energy density determines the expansion of the universe [6]. 
It is the most suitable model for our universe because it provides us with explanations of the 3 main cosmological phenomena:

| The existence and structure of CMBR | The accelerating expansion of the universe| Abundance of Hydrogen and Helium |

The abundance of Hydrogen (H) and Helium (He) is simply out of proportion with the number of observable stars in the universe, which suggests stars' nucleosynthesis is not the only source for the production of H and He. Thus this points to the Big Bang nucleosynthesis of H and He.

```python
'''Dictionary for omegas in the Lambda CDM model'''
Lambda_CDM={'H':70,'om_k':0,'om_M':0.3,'om_R':0,'om_lambda':0.7}
```

```python
'''Plotting the integrand for Lambda CDM model:'''
plt.figure()
a= np.linspace(0.01,1,100) #a cannot be 0 as division by zero 
plt.plot(a,Friedmann(a,Lambda_CDM['H'],Lambda_CDM['om_k'],Lambda_CDM['om_M'],Lambda_CDM['om_R'],Lambda_CDM['om_lambda']),label='$\Lambda$CDM Model')
#the area under is the age of the universe
plt.title('Plotting Integrand of the $\Lambda$CDM (Figure 1)')
plt.grid()
plt.legend()
```

- By plotting the integrand, we are able to see the function we are expected to solve by integrating for the age of the universe. Also, figure 1 indicates how well the function behaves as a model. Moreover, we can even compare the integrands of different models to predict which one should have the lowest value of $t$ (age of the universe). 

####  Age of the $\Lambda CDM$ Model universe :
#### Using The Trapezoidal Method to Integrate

The Trapezoidal rule is a method for evaluating definite integrals. 
A problem which arises when we try to find the age of the universe is that the equation cannot handle with a = 0 since it will become a division by zero. 

```python
'''A function to find the age of the universe using Trapezoidal method of integration'''
def time_trapz(a,H,om_k,om_M,om_R,om_lambda):
    time = np.trapz(Friedmann(a,H,om_k,om_M,om_R,om_lambda),a)
    age=(time*Mpc)/(km*Gyr) 
    return age  #returns the age of universe in gigayear 
#A function to find the age of the universe using Simpson's method of integration
def time_simps(a,H,om_k,om_M,om_R,om_lambda):
    time = integrate.simps(Friedmann(a,H,om_k,om_M,om_R,om_lambda),a)
    age=(time*Mpc)/(km*Gyr) 
    return age
```

```python
'''Calculating the age of Lambda CDM model: By using Trapezoid method to integrate'''
a = np.linspace(0.01,1,100)
age_CDM=time_trapz(a,Lambda_CDM['H'],Lambda_CDM['om_k'],Lambda_CDM['om_M'],Lambda_CDM['om_R'],Lambda_CDM['om_lambda'])
print('Age of Lambda CDM universe in Gyr (Trapezoid) =',age_CDM) #age in Gigayear
```

#### Reduce the starting value of $a$ to approach 0 for a more precise value of the age of the universe $t$

```python
'''Integrating with a approaching 0:'''
a= np.linspace(0.1e-27,1,100) #a cannot be 0 as division by zero 
age_CDM=time_trapz(a,Lambda_CDM['H'],Lambda_CDM['om_k'],Lambda_CDM['om_M'],Lambda_CDM['om_R'],Lambda_CDM['om_lambda'])
print('Age of Lambda CDM universe in Gyr (Trapezoid) =',age_CDM) #age in Gigayear
```

- This value of $t$ in Gyr is very closed to actual observations found by plotting red-shift of galaxies vs distances to each galaxies. From which $t$ is found by calculating the gradient as $t$ = $\frac{1}{H_0}$
- Through testing, the model shows that the smaller value of $a$ is the more accurate the value of $t$. This implies that the closer the value of $a$  is to zero, the more precise the integration is. However, $a$ can never reach zero as the integrand of equation 1.1 will become a division by zero, which Python cannot handle. 
- This model shows that the value of t can be calculated to be about $t$ = 13.47216418440882, which is found by evaluating at $a$ = 0.1e-27 to 1. After this point, decreasing the value to $a$ = 0.1e-28 doesn't change the accuracy of the value of t. This shows the limit of where computing the age of the universe can be limited by machine precision. Hence from this point we'll be evaluating with smallest value of $a$ = 0.1e-27. 

#### Using Simpsons method to integrate

```python
'''Calculating the age of Lambda CDM model: By using Simpson's method to integrate'''
a= np.linspace(0.1e-27,1,100) #a cannot be 0 as division by zero 
age_simps=time_simps(a,Lambda_CDM['H'],Lambda_CDM['om_k'],Lambda_CDM['om_M'],Lambda_CDM['om_R'],Lambda_CDM['om_lambda'])
print('Age of Lambda CDM universe in Gyr (Simpson) =',age_simps) #age in Gigayear
print('Age difference in Gyr =',abs(age_simps-age_CDM))
```

- This shows that the Simpson's method gives a different value of the age of the universe. 
- The Simpson's rule requires an even number of intervals. The parameter 'even' in the Scipy.integrate module regulates how integration is performed. It will either use Trapezoid rule on the last interval or the first interval. It can also perform on the average of the two the results (first-last).  
- The integration using the Simpson's method gives a better approximation than the Trapezoidal rule because it uses parabolas to approximate each part of the curve instead of straight lines to model a curve.  

```python
'''Measure time taken to execute for age of Lambda_CDM universe'''
%timeit time_trapz(a,Lambda_CDM['H'],Lambda_CDM['om_k'],Lambda_CDM['om_M'],Lambda_CDM['om_R'],Lambda_CDM['om_lambda'])
%timeit time_simps(a,Lambda_CDM['H'],Lambda_CDM['om_k'],Lambda_CDM['om_M'],Lambda_CDM['om_R'],Lambda_CDM['om_lambda'])
```

- The Trapezoid method took less time to evaluate for the area under the function. The Trapezoid rule is just the average of the left-hand and right-hand Riemann Sum which results in an estimate that's more accurate than either sum alone. However, Simpson's rule is a weighted average. 
- Simpson's rule uses parabolas, thus it is exact for any quadratic polynomial but the points must be evenly spaced.

### $\S$1.2 Einstein-de Sitter [EdS ] $(\Omega_M = 1)$
<a id = 'section1.2'></a>

Einstein-de Sitter is a matter dominated Friedmann model of the universe. It has zero curvature ($k$=0) [4] and it suggests the universe will expand forever with optimal energy to escape to infinity. The model is proposed by Albert Einstein and Willem de Sitter in 1932. The EdS model became a standard model of the universe because of its simplicity and due to the lack of empirical evidence for a cosmological constant and spatial curvature at the time. [10] 

```python
'''Dictionary for omegas in the EdS model'''
EdS={'H':70,'om_k':0,'om_M':1,'om_R':0,'om_lambda':0}
```

```python
'''Plotting the integrand of the EdS model:'''
plt.figure()
a= np.linspace(0.1e-27,1,100) #a cannot be 0 as division by zero 
plt.plot(a,Friedmann(a,EdS['H'],EdS['om_k'],EdS['om_M'],EdS['om_R'],EdS['om_lambda']),label='EdS Model')
plt.grid()
plt.title('Plotting Integrand of the Einstein-de Sitter (Figure 2)')
plt.legend()
```

Furthermore, this suggests the model predicts the EdS universe has expanded less rapidly than the $\Lambda$CDM.

#### Age of the EdS Model universe :

```python
'''Calculating the age of EdS model: by using Trapezoid method to integrate'''
a= np.linspace(0.1e-27,1,100) #a cannot be 0 as division by zero 
age_EdS=time_trapz(a,EdS['H'],EdS['om_k'],EdS['om_M'],EdS['om_R'],EdS['om_lambda'])
print('Age of EdS universe in Gyr (Trapezoid) =',age_EdS)  #age in Gigayear

a= np.linspace(0.1e-27,1,100) #a cannot be 0 as division by zero 
age_simps=time_simps(a,EdS['H'],EdS['om_k'],EdS['om_M'],EdS['om_R'],EdS['om_lambda'])
print('Age of Lambda CDM universe in Gyr (Simpson) =',age_simps) #age in Gigayear
print('Age difference in Gyr =',abs(age_simps-age_EdS))
```

- The Closed model shows the universe expands up to a point. After 10 Gyr the expansion begins to plateau.
- However, by using the original equation we cannot integrate past the maximum scale factor point $a(t)$ to observe the 'Big Crunch'. Thus we must find a more suitable equation. 

For the Closed model universe, we need the Friedmann equation in this form:

Equation 2
$$ H_0 t_0 = \int_0^a \frac{da}{\bigg(\frac{\Omega_0}{a} + (1-\Omega_0)\bigg)^\frac{1}{2}} $$ (Ryden 6.16) 
where dividing through by $H_0$ gives the age of the universe.

Equation 2.1
$$ H_0 t_0 = \frac{\Omega_0}{2(\Omega_0-1)^\frac{3}{2}} \arccos(\frac{2-\Omega_0}{\Omega_0}) - \frac{1}{\Omega_0 -1} $$ (Ryden 6.43)

Equation 2.2
$$ a(\theta) = \frac{1}{2} \frac{\Omega_0}{\Omega_0-1} (1-\cos(\theta)$$
(Ryden 6.17)

Equation 2.3
$$ t(\theta) = \frac{1}{2H_0} \frac{\Omega_0}{(\Omega_0-1)^\frac{3}{2}} (\theta-\sin(\theta)$$ (Ryden 6.18)
[7] Where the parameter $\theta$ spans from 0 $\to$ $\pi$


These equations allow us to evaluate past the maximum point and find the time of the 'Big Crunch'. 

```python
H = 70 
H0 = H*Gyr * km/Mpc
'''Returns the present time of a close universe relative to the input Omega_0 '''
def time_(omega, theta):
    eq=(1/(2*H))*(omega/((omega-1)**(3/2)))*(theta-np.sin(theta))/(km*Gyr/Mpc)
    return eq
'''returns the scale factor a of a closed universe'''
def scale_a(omega, theta): 
    eq=(1/2)*(omega/(omega-1))*(1-np.cos(theta))
    return eq
'''returns the time of close universe'''
def close(omega):
    equation=((omega/(2*(omega-1)**(3/2)))*(np.arccos((2-omega)/omega)))-(1/(omega-1))
    return equation
```

```python
'''Plotting the scale factor Vs age of universe for different test values of omega_0'''
plt.figure()
omegas=[1.1,1.2,1.5,2,3.0]
time_close = []
for i in omegas:
    time_close.append(close(i)/H0)

theta = np.arange(0,2*np.pi,0.1)   
scale_factor = []  
time = []
closed = [1.1, 1.2, 1.5, 2.0,3.0]

j=0
for i in (closed):
    scale_factor.append( scale_a(i,theta) )
    time.append(time_( i,theta) )
    plt.plot(time[j], scale_factor[j],label=closed[j])
    j+=1
plt.xlabel('Time $t$ (Gyr)')
plt.ylabel('Scale Factor $a(t)$')
plt.title('Scale Factor $a(t)$ Vs Time $t$ for Closed universe with varying densities $\Omega_0$ (Figure 10)')
plt.grid()
plt.legend(title ='$\Omega_0$' ,loc='upper right')
```

- From the plots, we can see clearly that the larger the density parameter the quicker the collapse. This suggests the universe that contains more densities decelerates quicker, which then leads to the 'Big Crunch' sooner. This is due to the fact that as the density increases the gravitational forces pull the universe back together faster than the expansion rate of the universe.  
- In this model, we cannot evaluate for a flat universe as when we input $\Omega_0$ = 1.0 there would be a division by zero.

 
### $\S$ 2.5 Open Universe $\Omega_{0} < 1$
<a id = 'section2.5'></a>
Equation 3
$$H_0 t_0 = \frac{1}{1-\Omega_0} - \frac{\Omega_0}{2(1-\Omega_0)^\frac{3}{2}} \cosh^{-1} \bigg(\frac{2-\Omega_0}{\Omega_0}\bigg)$$
(Ryden 6.44) 

Equation 3.1
$$a(\eta) = \frac{1}{2} \frac{\Omega_0}{1-\Omega_0} \big(\cosh(\eta)-1\big)$$
(Ryden 6.20)

Equation 3.2
$$t(\eta) = \frac{1}{2H_0} \bigg(\frac{\Omega_0}{1-\Omega_0}\bigg)^\frac{3}{2} \big(\sinh(\eta)-\eta\big)$$ (Ryden 6.21) 
where $\eta$ runs from 0 $\to$ infinity [7]

```python
'''Returns the present age of the input Omega'''
def t0(omega):
    e= 1/(1-omega) - (omega/(2*(1-omega)**(3/2)) * np.arccosh((2-omega)/omega))
    return e
'''Returns the scale factor of open universe relative to input Omega'''
def open_scale(omega,eta):
    b=(1/2)*(omega/(1-omega))*(np.cosh(eta)-1)
    return b
'''Returns the age of an open universe'''
def open_time(omega,eta):
    c= (1/2)*(1/H0)*(omega/((1-omega)**(3/2)))*(np.sinh(eta)-eta)
    return c
```

```python
'''Plotting the age of the universe against the scale factor for open universes'''
plt.figure()
eta= np.arange(0,10,0.01)    
open_scalefactor=[]
open_ages=[]
ope = [0.1,0.3,0.5,0.7,0.99]   #list of different Omega_0 for Open universes
ages=[]
for i in ope:
    ages.append(t0(i)/(H0)) #appends the age of the universe in Gyr 
j=0
for i in ope:
    open_scalefactor.append(open_scale(i,eta))      
    open_ages.append(open_time(i,eta))
    plt.plot(open_ages[j],open_scalefactor[j],label=ope[j])
    j+=1
plt.grid()
plt.xlim(0,100)
plt.ylim(0,3)
plt.legend(title='$\Omega_0$:')
plt.xlabel('Time $t$ (Gyr)')
plt.ylabel('Scale Factor $a(t)$')
plt.title('Scale Factor $a(t)$ Vs different $\Omega_0$ for Open Universe (Figure 11)')
```

- When the density parameter $\Omega_0$ is below 1 the scale factor increases indefinitely and thus the universe expands forever. To confirm and see this more easily we can increase the $x$-limit (time), which would show the graphs with steeper gradients no matter how much we increase the time limit to find where it may come back on itself. 
- By making adjustments to the step size eta the line becomes more or less steep. That is if the step size is decreased the lines become less smooth. Also, the same can happen if we vary the actual range of eta. However, if eta is evaluated from 0 $\to$ 1 we will not be able to see the curves we expected because the range is too small to see any changes.
